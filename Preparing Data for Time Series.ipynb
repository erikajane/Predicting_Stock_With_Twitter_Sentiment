```python
# Import necessary libraries

import pandas as pd
```

```python
# Import and preview daily twitter senitment data 

daily_sentiment = pd.read_csv('daily_sentiment_scores.csv')
daily_sentiment.head()
```

```python
# Import and preview Tesla daily stock data

stock_prices = pd.read_csv('TSLA.csv')
stock_prices.head()
```

```python
# Check daily_sentiment datatypes

daily_sentiment.info()
```

```python
# Change daily_sentiment 'date_time' column into datetime and set as index

daily_sentiment.set_index(pd.to_datetime(daily_sentiment.date_time), inplace = True)
daily_sentiment = daily_sentiment.drop(columns = ['date_time'])

daily_sentiment.head()
```

```python
# Check stock_prices datatypes

stock_prices.info()
```

```python
# Change stock_prices 'Date' column into datetime and set as index

stock_prices.set_index(pd.to_datetime(stock_prices.Date), inplace = True)
stock_prices = stock_prices.drop(columns = ['Date'])
stock_prices.tail()
```

```python
# Create column to view the change in open prices from prior days close
# Create a column that labels whether the open-close difference was positive or negative

stock_prices['open_close_diff'] = stock_prices.Open - stock_prices.Close.shift(1)
stock_prices['pos_neg'] = stock_prices['pos_neg'] = stock_prices.open_close_diff.apply(lambda x: 1 if (x > 0) else (-1 if x < 0 else 0))
stock_prices
```

```python
# Upsample so you have columns for everyday

stock_prices = stock_prices.resample('D').asfreq()
stock_prices
```

```python
# Merge both dataframe

daily = pd.merge(stock_prices, daily_sentiment, how='inner', left_index=True, right_index=True)
daily.head()
```

```python
daily
```

```python
# Drop unnecessary columns

daily = daily.drop(columns = ['Open','High', 'Low', 'Close', 'Adj Close', 'Volume'])
daily.head()
```

## Fix Friday Sentiment Scores

#### Average Friday, Saturday and Sunday sentiment scores and assign new sentiment score for Friday

```python
# Copy the daily dataframe

new_df = daily.copy()
new_df.head(7)
```

```python
# Create a dataframe with only Friday's

fridays = new_df[new_df.index.day_name() == 'Friday'].reset_index()
fridays = fridays.rename(columns={"index": "date"})
fridays.columns = ['friday_' + str(col) for col in fridays.columns]
fridays.head()
```

```python
# Create a dataframe with only Saturday's

saturdays = new_df[new_df.index.day_name() == 'Saturday'].reset_index()
saturdays = saturdays.rename(columns={"index": "date"})
saturdays.columns = ['saturday_' + str(col) for col in saturdays.columns]
saturdays.head()
```

```python
# Create a dataframe with only Sunday's

sundays = new_df[new_df.index.day_name() == 'Sunday'].reset_index()
sundays = sundays.rename(columns={"index": "date"})
sundays.columns = ['sunday_' + str(col) for col in sundays.columns]
sundays.head()
```

```python
# Merge frudays, saturdays, and sundays into one dataframe

weekends = pd.concat([fridays, saturdays, sundays], axis=1, sort=False).reindex()
weekends.head()
```

```python
# For each type of sentiment score, find the average for the weekend

weekends['sentiment_1'] = (weekends['friday_sentiment_1'] + weekends['saturday_sentiment_1'] + weekends['sunday_sentiment_1'])/3
weekends['sentiment_2'] = (weekends['friday_sentiment_2'] + weekends['saturday_sentiment_2'] + weekends['sunday_sentiment_2'])/3
weekends['s1_no_0'] =  (weekends['friday_s1_no_0'] + weekends['saturday_s1_no_0'] + weekends['sunday_s1_no_0'])/3
weekends['s2_no_0'] =  (weekends['friday_s2_no_0'] + weekends['saturday_s2_no_0'] + weekends['sunday_s2_no_0'])/3
weekends['s1_scaled'] =  (weekends['friday_s1_scaled'] + weekends['saturday_s1_scaled'] + weekends['sunday_s1_scaled'])/3
weekends['s2_scaled'] =  (weekends['friday_s2_scaled'] + weekends['saturday_s2_scaled'] + weekends['sunday_s2_scaled'])/3
weekends = weekends[['friday_date', 'saturday_date', 'sunday_date','sentiment_1', 'sentiment_2', 's1_no_0', 's2_no_0', 's1_scaled', 's2_scaled']]
weekends.head()
```

```python
# Replace the Friday sentiment scores in the daily dataframewith the new sentiment scores from weekends dataframe

for i in range(len(daily)):
    for j in range(len(weekends)):
        if daily.index[i] == weekends.friday_date[j]:
            daily.sentiment_1[i] = weekends.sentiment_1[j]
            daily.sentiment_2[i] = weekends.sentiment_2[j]
            daily.s1_no_0[i] = weekends.s1_no_0[j]
            daily.s2_no_0[i] = weekends.s2_no_0[j]
            daily.s1_scaled[i] = weekends.s1_scaled[j]
            daily.s2_scaled[i] = weekends.s2_scaled[j]
daily.head(7)
```

```python
# Check for null values (Saturday's and Sunday's)

daily.isnull().sum()
```

```python
# Drop weekend rows

daily = daily.dropna()
```
